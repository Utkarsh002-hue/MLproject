```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(pd.read_csv("data/dataset.csv"))
df.head(10)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Unnamed: 0','uid','asin','title'],inplace=True)
```

```python
df.head(10)
```

```python
df.describe
```

```python
duplicate_series = df.duplicated(['stars','reviews','category','isBestSeller','boughtInLastMonth'],keep='first')
```

```python
duplicate_series
```

```python
data = df[~duplicate_series]
data
```

```python
data.info()
```

```python
numerical_data = data.select_dtypes(include = [np.number])
categorical_data = data.select_dtypes(exclude = [np.number])
print(numerical_data)
print(categorical_data)
```

```python
from sklearn.preprocessing import LabelEncoder
label_model = LabelEncoder()
label_model.fit(categorical_data['category'])
categorical_data['category'] = label_model.transform(categorical_data['category'])
label_model.fit(categorical_data['isBestSeller'])
categorical_data['isBestSeller'] = label_model.transform(categorical_data['isBestSeller'])
categorical_data
```

```python
data = pd.concat([numerical_data, categorical_data],axis = 1, join = 'inner')
data.head(10)
```

## Feature Engineering : 

```python
data['fair_rated'] = np.where(data['reviews'] >= data['reviews'].mean(),1,0)
data.head(10)
```
